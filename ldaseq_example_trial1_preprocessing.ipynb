```
## preparation for text preprocessing
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))
from nltk.stem import PorterStemmer
ps = PorterStemmer()
```

```
#loading corpus and preprocessing
corpus_list= []
for entry in entries:
    filepath = "/".join(["/content/drive/My Drive/words/sample", entry, "*.txt"])
    data_files = glob.glob(filepath)
    for data_file in data_files:
      text = open(data_file,'r').read().lower().replace('"', "")
      filtered = [word for word in text.split() if not word in stop_words] ##revome stop words
      corpus = [ps.stem(word) for word in filtered] ##stem the word
      corpus_list += [corpus]
len(corpus_list)



```

```
## find distinct words
word_list = [word.strip() for corpus in corpus_list for word in corpus]
s = set(word_list)
distinct_list = sorted(list(s))

## create a dictionary for distinct word
word_dict = dict(enumerate(distinct_list))
word_dict
```

```
## transform corpus into blei form
new_corpus = [" ".join(corpus) for corpus in corpus_list]

##time slice
time_slice = [438, 430, 456]
```

```
ldaseq = ldaseqmodel.LdaSeqModel(corpus = new_corpus, id2word=word_dict, time_slice= time_slice, num_topics= 5)
```
